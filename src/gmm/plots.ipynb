```python
import matplotlib.pyplot as plt

# For consistent, publication-quality aesthetics
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 14,
    "axes.titlesize": 15,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "figure.dpi": 300,
    "figure.figsize": (6, 4),
    "lines.linewidth": 2,
    "lines.markersize": 6,
})

# Example data
benchmarks = ["GSM8k", "HellaSwag", "MMLU"]
x = range(len(benchmarks))

baseline = [45.2, 63.5, 52.1]
focal = [47.6, 65.1, 54.8]
tce = [48.3, 67.0, 56.3]

# Create the plot
fig, ax = plt.subplots()

ax.plot(x, baseline, marker='o', label='Baseline (CE)')
ax.plot(x, focal, marker='s', label='Focal Loss')
ax.plot(x, tce, marker='^', label='TCE')

# Customize ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(benchmarks)
ax.set_ylabel("Accuracy (%)")
ax.set_xlabel("Benchmark")
ax.set_title("Performance Comparison Across Benchmarks")
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(frameon=False)

# Tight layout for better spacing
plt.tight_layout()
plt.show()
```
